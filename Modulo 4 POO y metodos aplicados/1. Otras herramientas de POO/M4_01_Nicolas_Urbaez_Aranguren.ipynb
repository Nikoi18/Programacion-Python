**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
class Vehiculo():

    def __init__(self, marca=None, modelo=None):
        self.marca=marca
        self.modelo=modelo

    def repostar(self):
        return "Este vehiculo tiene que repostar gasolina"

    def __str__(self):
        return "Marca: {} Modelo: {}".format(self.marca, self.modelo)

class VElectricos():

    def __init__(self, marca=None, modelo=None, autonomia=None):
        self.marca=marca
        self.modelo=modelo
        self.autonomia=autonomia
    
    def repostar(self):
        return "Este Vehiculo tiene que repostar Electricidad"

    def __str__(self):
        return "Marca: {}, Modelo: {}, Autonomia: {}".format(self.marca, self.modelo, self.autonomia)

class BicicletaElectrica(VElectricos, Vehiculo):

    def __init__(self, marca=None, modelo=None, autonomia=None):
        VElectricos.__init__(self, marca, modelo, autonomia)
        Vehiculo.__init__(self,marca, modelo)

    def __str__(self):
        return VElectricos.__str__(self)

class Quad(Vehiculo, VElectricos):

    def __init__(self, marca=None, modelo=None, autonomia=None):
        Vehiculo.__init__(self, marca, modelo)
        VElectricos.__init__(self, marca, modelo, autonomia)

    def __str__(self):
        return Vehiculo.__str__(self)

def estado(respotaje):
    print(respotaje.repostar())


v2=BicicletaElectrica(marca="toyota", modelo="v2.5", autonomia="90%")
print(v2)
estado(v2)
v3=Quad(marca="MS2", modelo="monta")
print(v3)
estado(v3)    
```

**2) Ejercicio de herencia**
* Crea la superclase Personal_Universitario con un único atributo que sea un diccionario que incluya: "id", "nombre" y "email"
* Crear la clases Oficina, Profesor y Alumno. Todas ellas heredaran de Personal_Universitario y heredaran sus datos
* En el caso de Oficina, debe añadir a su diccionario de datos el dato "Puesto"
* En el caso de Profesor, debe añadir a su diccionario de datos el dato "Especialización"
* En el caso de Alumno, debe añadir a su diccionario de datos el dato "CreditosAprobados" (integer)
* Crear métodos que sirvan para mostrar la información de cada clase
* Crear objetos de todas las clases
* Mostrar la información de esos objetos y comprobar que sale lo que tiene que salir

```python
class Personal_Universitario:

    def __init__(self, datos=None):
        self.datos={}
        self.datos["id"]=datos["id"]
        self.datos["nombre"]=datos["nombre"]
        self.datos["email"]=datos["email"]

    def __str__(self):
        return "Id: {} \nNombre: {} \nEmail: {}".format(self.datos["id"], self.datos["nombre"], self.datos["email"], end="")
        
class Oficina(Personal_Universitario):

    def __init__(self, datos, puesto):
        super().__init__(datos)
        self.datos["puesto"]=puesto

    def __str__(self):
        return super().__str__() + "\nPuesto: {}".format(self.datos["puesto"])

class Profesor(Personal_Universitario):

    def __init__(self, datos, especializacion):
        super().__init__(datos)
        self.datos["especializacion"]=especializacion

    def __str__(self):
        return super().__str__()+ "\nEspecializacion: {}".format(self.datos["especializacion"])

class Alumno(Personal_Universitario):

    def __init__(self, datos, creditosAprobados):
        super().__init__(datos)
        self.datos["creditosAprobados"]=creditosAprobados

    def __str__(self):
        return super().__str__()+ "\nCreditos Aprobados: {}".format(self.datos["creditosAprobados"])

personal1={"id":"0001", "nombre":"Nicolas", "email":"aran.nick15@gmail.com"}
ofi=Oficina(personal1, "fijo")
print(ofi)

personal2={"id":"0002", "nombre":"Alberto","email":"aran.nick15@gmail.com"}
ofi2=Profesor(personal2, "programador")
print(ofi2)

personal3={"id":"0003","nombre":"Milagros","email":"mili.janina15@gmail.com"}
ofi3=Alumno(personal3,20)
print(ofi3)
```

```python
class Liceo:

    def __init__(self, nombre, ubicacion):
        self.nombre=nombre
        self.ubicacion=ubicacion

    def __str__(self):
        return "Nombre del Liceo: {} \nUbicacion: {}".format(self.nombre, self.ubicacion)

class Alumnos(Liceo):

    def __init__(self, nombre, ubicacion, nombreAlumno, apellidoAlumno):
        super().__init__(nombre, ubicacion)
        self.nombreAlumno=nombreAlumno
        self.apellidoAlumno=apellidoAlumno

    def __str__(self):
        return super().__str__()+"\nNombre Alumno: {} \nApellido Alumno: {}".format(self.nombreAlumno, self.apellidoAlumno)

    def turno(self):
        return "dia"

class Notas(Alumnos):

    def __init__(self, nombre=None, ubicacion=None, nombreAlumno=None, apellidoAlumno=None, matematica=None, castellano=None, deporte=None):
        super().__init__(nombre, ubicacion, nombreAlumno, apellidoAlumno)
        self.matematica=matematica
        self.castellano=castellano
        self.deporte=deporte

    def __str__(self):
        return super().__str__()+"\nNotas: \nMatematica: {} \nCastellano: {} \nDeporte: {}".format(self.matematica, self.castellano, self.deporte)

class Profesor(Liceo):

    def __init__(self, nombre, ubicacion, nombreProfesor, materia):
        super().__init__(nombre, ubicacion)
        self.nombreProfesor=nombreProfesor
        self.materia=materia

    def __str__(self):
        return super().__str__()+"\nNombre Profesor: {} \nMateria: {}".format(self.nombreProfesor, self.materia)

    def turno(self):
        return "Tarde"

class Trabajadores(Liceo):

    def __init__(self, nombre, ubicacion, nombreTrabajador, puesto):
        super().__init__(nombre, ubicacion)
        self.nombreTrabajador=nombreTrabajador
        self.puesto=puesto

    def __str__(self):
        return super().__str__()+"\nNombre Trabajador: {} \nPuesto: {}".format(self.nombreTrabajador, self.puesto)

    def turno(self):
        return "Noche\n"

def turnos(self):
    print(self.turno())

#liceo= Liceo(nombre="Ezequiel Zamora", ubicacion="Punta de Mata")
#print(liceo)

alumno= Alumnos(nombre="Ezequiel Zamora", ubicacion="Punta de Mata", nombreAlumno="Nicolas", apellidoAlumno="Urbaez")
print(alumno)
turnos(alumno)

profe= Profesor(nombre="Ezequiel Zamora", ubicacion="Punta de Mata", nombreProfesor="Juan", materia="Matematica")
print(profe)
turnos(profe)
tra= Trabajadores(nombre="Ezequiel Zamora", ubicacion="Punta de Mata", nombreTrabajador="Diablo", puesto="Dueño del liceo")
print (tra)
turnos(tra)
no = Notas(nombre="Liceo", ubicacion="PDM", nombreAlumno="Nicolas", apellidoAlumno="Urbaez", matematica="05", castellano="08", deporte="10")
print (no)
turnos(no)
```
